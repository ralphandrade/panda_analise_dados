```python
import pandas as pd

# lê os dados da dataset

tabela = pd.read_csv('cancelamentos.csv')


# mostra o dataset na tela
display(tabela)
```

agora vo eliminar da consulta dados desnecessários. vou eliminar o id do usuário.

```python
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```

vamos descobrir quantas linhas de informação vazias temos.

```python
display(tabela.info())
```

vamos eliminar as células vazias da nossa consulta

```python
tabela = tabela.dropna()
display(tabela.info())
```

vamos contar quantos usuarios cancelaram e quantos não cancelaram

```python
display(tabela['cancelou'].value_counts())
```

Traduzindo para porcentagem

```python
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

Próxima análise de dados vai ser pra verificar a duração dos contratos

```python
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:.2%}'.format))
```

vamos agrupar as informações da duração de contrato


```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

vamos eliminar os dados do contrato mensal

```python
tabela = tabela[tabela['duracao_contrato']!='Monthly']
display(tabela['cancelou'].value_counts)
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

vamos dar uma olhada nos tipos de assinatura dos planos:

```python
display(tabela['assinatura'].value_counts(normalize=True).map('{:.2%}'.format))
display(tabela.groupby('assinatura').mean(numeric_only=True))
```

Vamos passar analisar os dados das assinaturas olhando os gráficos de cada coluna

```python
import plotly.express as px
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color='cancelou', width=600)
    grafico.show()
```

achamos as anomalias:ligações para call center e dias de atraso. Os clientes com mais de 20 dias de atraso cancelaram as assinaturas, enquanto que os clientes que fizeram mais de 5 ligações para o call center, todos cancelaram as suas assinaturas

```python
tabela = tabela[tabela['ligacoes_callcenter'] < 5]
tabela = tabela[tabela['dias_atraso'] <= 20]

display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))


```
